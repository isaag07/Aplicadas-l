```python
# @title Gráfico para función $f(x)=x^n$
valor_de_n = 8 # @param {"type":"slider","min":1,"max":10,"step":1}
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-100,100, 1000)
y = x**valor_de_n
plt.plot(x,y)


plt.title('Funciones $x^n$ para n ∈ [1, 10]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')
plt.show()




```

```python
import numpy as np
import matplotlib.pyplot as plt

v0 = 20
g = 9.8

angulos=[]
for i in range(5):
  angulo = int(input("Escribe el angulo de lanzamiento: "))
  angulos.append(angulo)


# Simular y graficar cada ángulo
for i in angulos:
    theta = np.radians(i)
    t_vuelo = (2 * v0 * np.sin(theta)) / g  # tiempo total de vuelo
    t = np.linspace(0, t_vuelo, 100)  # vector de tiempos

    # Posiciones x y y
    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2

    # Graficar trayectoria
    plt.plot(x, y, label=f'{i}°')

# Personalización del gráfico
plt.title('Movimiento parabólico para diferentes ángulos')
plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid(True)
plt.show()
```
