```python
import numpy as np
import matplotlib.pyplot as plt

def fibonacci(n):
    """ Genera los primeros n términos de la serie de Fibonacci. """
    if n < 1:
        return []

    fib_series = np.zeros(n, dtype=int)
    fib_series[0] = 1
    if n > 1:
        fib_series[1] = 1
        for i in range(2, n):
            fib_series[i] = fib_series[i-1] + fib_series[i-2]

    return fib_series

def plot_fibonacci(n):
    """ Genera y grafica los primeros n términos de la serie de Fibonacci. """
    fib_values = fibonacci(n)
    indices = np.arange(1, n+1)

    plt.figure(figsize=(8, 5))
    plt.plot(indices, fib_values, marker='o', linestyle='-', color='b', label="Fibonacci")
    plt.xlabel("Índice n")
    plt.ylabel("Valor de Fibonacci")
    plt.title(f"Primeros {n} términos de la serie de Fibonacci")
    plt.grid()
    plt.legend()
    plt.show()

# Ejemplo de uso:
n = 10  # Cambia este valor para ver más términos
plot_fibonacci(n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def es_primo(n):
    """Determina si un número es primo."""
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def primeros_primos(n):
    """Genera los primeros n números primos."""
    primos = []
    num = 2
    while len(primos) < n:
        if es_primo(num):
            primos.append(num)
        num += 1
    return np.array(primos)

def graficar_primos(n):
    """Grafica la distribución de los primeros n números primos."""
    primos = primeros_primos(n)
    indices = np.arange(1, n + 1)

    plt.figure(figsize=(8, 5))
    plt.scatter(indices, primos, color='red', label="Números primos")
    plt.xlabel("Índice")
    plt.ylabel("Valor del número primo")
    plt.title(f"Distribución de los primeros {n} números primos")
    plt.grid()
    plt.legend()
    plt.show()

# Ejemplo de uso:
n = 20  # Cambia este valor para ver más términos
graficar_primos(n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cuadrados(n):
    """Genera los primeros n números cuadrados."""
    return np.array([i**2 for i in range(1, n+1)])

def graficar_cuadrados(n):
    """Grafica la sucesión de números cuadrados."""
    valores = cuadrados(n)
    indices = np.arange(1, n+1)

    plt.figure(figsize=(8, 5))
    plt.plot(indices, valores, marker='o', linestyle='-', color='g', label="Números cuadrados")
    plt.xlabel("Índice n")
    plt.ylabel("Valor de n^2")
    plt.title(f"Primeros {n} términos de la sucesión de cuadrados")
    plt.grid()
    plt.legend()
    plt.show()

# Ejemplo de uso:
n = 15  # Cambia este valor para ver más términos
graficar_cuadrados(n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def suma_numeros(n):
    """Calcula la suma de los primeros n números naturales usando la fórmula de Riemann."""
    suma_formula = (n * (n + 1)) // 2  # Fórmula de la suma de los primeros n números
    return suma_formula

def graficar_suma(n):
    """Grafica la suma de los primeros n números naturales como una aproximación a la integral."""
    x = np.arange(1, n+1)
    y = np.cumsum(x)  # Suma acumulativa

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, marker='o', linestyle='-', color='b', label="Suma acumulada")
    plt.xlabel("n")
    plt.ylabel("Suma de los primeros n números")
    plt.title(f"Suma de los primeros {n} números naturales (Riemann)")
    plt.grid()
    plt.legend()
    plt.show()

# Ejemplo de uso:
n = 20  # Cambia este valor para calcular otros términos
print(f"Suma de los primeros {n} números naturales: {suma_numeros(n)}")
graficar_suma(n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_exponencial(n):
    """Calcula la aproximación de e con la fórmula (1 + 1/n)^n para una lista de valores."""
    valores = np.array(n)
    resultados = (1 + 1/valores)**valores
    return resultados

def graficar_exponencial(n):
    """Grafica la aproximación de e y la recta y = e."""
    valores = np.arange(1, n+1)
    resultados = calcular_exponencial(valores)
    e_real = np.e

    plt.figure(figsize=(8, 5))
    plt.plot(valores, resultados, marker='o', linestyle='-', color='g', label="Aproximación de e")
    plt.axhline(y=e_real, color='r', linestyle='--', label="y = e")
    plt.xlabel("n")
    plt.ylabel(r"$\left(1 + \frac{1}{n} \right)^n$")
    plt.title("Aproximación de e usando la sucesión (1 + 1/n)^n")
    plt.grid()
    plt.legend()
    plt.show()

# Ejemplo de uso:
n = 50  # Cambia este valor para calcular más puntos
graficar_exponencial(n)
```

```python
def calcular_nota_final(nota1, nota2, objetivo=4.2):
    """Calcula la nota necesaria en el tercer parcial para alcanzar un promedio de 4.2."""
    peso1, peso2, peso3 = 0.3, 0.4, 0.3
    nota_necesaria = (objetivo - (nota1 * peso1 + nota2 * peso2)) / peso3
    return max(1.0, min(5.0, nota_necesaria))  # La nota debe estar entre 1 y 5

# Ejemplo de uso:
nota1 = 4.0
nota2 = 3.8
nota_final = calcular_nota_final(nota1, nota2)
print(f"Para obtener un promedio de 4.2, se necesita sacar {nota_final:.2f} en el tercer parcial.")
```
