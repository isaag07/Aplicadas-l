#Laboratorio 1 - Cajero Automático

```python
import numpy as np

def generar_saldo_inicial():
    return np.random.randint(1000, 15000)

saldo = generar_saldo_inicial()
saldo_otro_usuario = generar_saldo_inicial()

def registrar_transaccion_en_archivo(nombre, tipo, monto):
    with open('transacciones.txt', 'a') as file:
        file.write(f"{nombre},{tipo},{monto:.2f}\n")

def imprimir_historial_transacciones():
    try:
        print("\nHistorial de transacciones:")
        with open('transacciones.txt', 'r') as file:
            datos = file.readlines()
            if not datos:
                print("No hay transacciones registradas aún.")
                return
            for linea in datos:
                info = linea.strip().split(',')
                if len(info) == 3:
                    nombre, tipo, monto = info
                    print(f"{nombre} - {tipo}: ${monto}")
    except FileNotFoundError:
        print("No hay transacciones registradas aún.")

def consultar_saldo():
    print("\nTu saldo actual es:")
    print(f"${saldo:.2f}")

def depositar(nombre):
    global saldo
    monto = float(input("\n¿Cuánto deseas depositar?: "))
    if monto <= 0:
        print("El monto debe ser mayor que cero.")
        return
    saldo += monto
    registrar_transaccion_en_archivo(nombre, "Depósito", monto)
    print("\nDepósito exitoso.")
    consultar_saldo()

def retirar(nombre):
    global saldo
    monto = float(input("\n¿Cuánto deseas retirar?: "))
    if monto <= 0:
        print("El monto debe ser mayor que cero.")
        return
    if monto > saldo:
        print("No tienes suficiente saldo.")
        return
    saldo -= monto
    registrar_transaccion_en_archivo(nombre, "Retiro", monto)
    print("\nRetiro exitoso.")
    consultar_saldo()

def transferir(nombre):
    global saldo, saldo_otro_usuario
    monto = float(input("\n¿Cuánto deseas transferir?: "))
    if monto <= 0:
        print("El monto debe ser mayor que cero.")
        return
    if monto > saldo:
        print("No tienes suficiente saldo para esta transferencia.")
        return
    saldo -= monto
    saldo_otro_usuario += monto
    registrar_transaccion_en_archivo(nombre, "Transferencia", monto)
    print("\nTransferencia exitosa.")
    consultar_saldo()

def ingresar_contrasena(documento, contrasena):
    intentos = 3
    while intentos > 0:
        entrada_doc = int(input("\nIngresa tu documento: "))
        entrada_contrasena = int(input("Ingresa tu contraseña: "))
        if entrada_doc == documento and entrada_contrasena == contrasena:
            print("\nAcceso concedido. Bienvenido.")
            return True
        intentos -= 1
        print(f"Credenciales incorrectas. Intentos restantes: {intentos}")
    print("Acceso denegado.")
    return False

def cajero_automatico():
    print("\nBienvenido al cajero automático ICESI")
    print("--------------------------------------")
    nombre = input("¿Cómo te llamas?: ").title()
    print(f"\nRegistrando a {nombre}.")
    documento = int(input("Ingresa tu documento: "))
    contrasena = int(input("Crea una contraseña numérica: "))
    print("\nRegistro exitoso.")

    if not ingresar_contrasena(documento, contrasena):
        return

    while True:
        print("\nMenú de opciones:")
        print("1. Consultar saldo")
        print("2. Depositar dinero")
        print("3. Retirar dinero")
        print("4. Transferir dinero")
        print("5. Ver historial de transacciones")
        print("6. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            consultar_saldo()
        elif opcion == "2":
            depositar(nombre)
        elif opcion == "3":
            retirar(nombre)
        elif opcion == "4":
            transferir(nombre)
        elif opcion == "5":
            imprimir_historial_transacciones()
        elif opcion == "6":
            print("\nGracias por usar el cajero automático. Hasta pronto.")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

cajero_automatico()
```
